##Importando Bibliotecas necessárias

```python
import vedo
from totalsegmentator.python_api import totalsegmentator
import nibabel as nib
import numpy as np
import os
```

##Preparando o Ambiente para o vedo

```python
vedo.settings.default_backend = 'k3d'
```

##Serviço de Segmentação(SegmentationService)

```python
def segmentate(image:str,output_name:str):
    totalsegmentator(f"{image}",f"{output_name}",fastest=True)
```

##Serviço para processar dados dos segmentos(ProcessSegmentService)

```python
def process_segment(input_image):
    segmentation = nib.load(f"{input_image}")
    segmentation_data = segmentation.get_fdata()
    return segmentation_data if np.sum(segmentation_data) != 0 else Exception("Image is null / Segment not found on the image")

```

##Serviço para visualização de segmentos processados(ShowSegmentsService)

```python
def show_segments(input_image):
    volume = vedo.Volume(input_image)
    mesh = volume.isosurface(value=0.5).c('red')
    vedo.show(mesh, __doc__, axes=1)
```

##Aplicação

```python
input_image = '/content/drive/MyDrive/Medis_Images/example_ct_sm.nii.gz'
output_image = '/content/drive/MyDrive/Medis_Results'
segmentate(input_image,output_image)
```

##Segmentos Criados

##Processando Segmentos

```python
segment = '/content/drive/MyDrive/Medis_Results/stomach.nii.gz'
processed_segment = process_segment(segment)
```

##Visualizando segmentos processados

```python
show_segments(processed_segment)
```
